파일네임 리스트, 얘를 스티쳐에 줄것입니다.  
for문에서 img_names의 각각을 name에 저장합니다.  
name을 불러와 img에 저장을 해 줍니다.  
그리고 imgs 리스트에 하나씩 추가하죠

```python
img_names = ['img1.jpg', 'img2.jpg', 'img3.jpg']


imgs = []

for name in img_names:#img_names의 각각을 name에 저장

    img = cv2.imread(name) #name을 불러와 img에 저장



    if img is None:

        print('Image load failed!')

        sys.exit()


    imgs.append(img) 
```

```python
plt.subplot(131),plt.axis('off'),plt.imshow(imgs[0])
plt.subplot(132),plt.axis('off'),plt.imshow(imgs[1])
plt.subplot(133),plt.axis('off'),plt.imshow(imgs[2])
```

# 스티처 클래스 생성

```python
stitcher = cv2.Stitcher_create()
```

# 클래스에서 스티치 함수호출을 통해 결과영상을 받습니다. 

```python
status, dst = stitcher.stitch(imgs)
```

```python
if status != cv2.Stitcher_OK:

    print('Stitch failed!')

    sys.exit()
```

```python
cv2.imwrite('output.jpg', dst)
```

```python
plt.axis('off'),plt.imshow(dst)
```

```python
cv2.namedWindow('dst', cv2.WINDOW_NORMAL)

#화면이 클 가능성이있으니 normal로 !

cv2.imshow('dst', dst)

cv2.waitKey()

cv2.destroyAllWindows()

​
```
