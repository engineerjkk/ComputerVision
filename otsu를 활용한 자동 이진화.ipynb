# 오츠를 활용한 전체이진화 실습  
  
두개의 클래스를 나눠놓고 각각의 분산값을 계산합니다. 합을 계산해서 Within-Class Variance로 했습니다.  
분산을 계산하고, 평균을 계산하고, 각각의 데이터와 평균의 차이의 제곱을 평균냅니다. 즉, 평균 여러번하고 T값이 바뀔 때마다 계속하니까 시간이 오래걸린다고 생각하지만 사실 이전의 값을 받아서 하는 특성을 갖기때문에 느리지 않습니다.

```python
import sys
import numpy as np
import cv2
import matplotlib.pyplot as plt

!pip install opencv-python
```

```python
src=cv2.imread('rice.png',cv2.IMREAD_GRAYSCALE)
```

```python
if src is None:
    print('image load failed!')
    sys.exit()
```

오츠를 활용한 스레시홀드 함수는 두개의 인자를 반환합니다. threshold값과 출력영상입니다. 스레시홀드 값은 0을, 최대값은 255을 줍니다. 마지막 파라미터는 바이너리와 오츠를 OR연산으로 묶어서 해주시면 됩니다. 하지만 바이너리를 안쓰셔도 있다고 가정하기 때문에 안써도 되긴합니다.

```python
th,dst=cv2.threshold(src,0,255,cv2.THRESH_BINARY|cv2.THRESH_OTSU)
```

```python
print("otsu's threshold",th)
```

```python
plt.subplot(121),plt.axis('off'),plt.imshow(src,cmap='gray')
plt.subplot(122),plt.axis('off'),plt.imshow(dst,cmap='gray')
```

```python
cv2.imshow('src',src)
cv2.imshow('dst',dst)
cv2.waitKey()
cv2.destroyAllWindow()
```
