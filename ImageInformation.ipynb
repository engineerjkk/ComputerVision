```python
img1 = cv2.imread('jkk_burger.jpg', cv2.IMREAD_GRAYSCALE)

#cat.bmp 파일을 그레이스케일로 불러와서 img1에다가 넣는다.

img2 = cv2.imread('jkk_burger.jpg', cv2.IMREAD_COLOR)
```

```python
if img1 is None or img2 is None:

    print('Image load failed!')

    sys.exit()
```

```python
print(type(img1))

print(img1.shape)#2차원

print(img2.shape)#3차원으로 출력된다.

print(img1.dtype)

print(img2.dtype)#똑같이 uint8로 출력된다.
```

```python
# 영상의 속성 참조

print('type(img1):', type(img1))

print('img1.shape:', img1.shape)

print('img2.shape:', img2.shape)

print('img1.dtype:', img1.dtype)
```

```python
# 영상의 크기 참조

h, w = img2.shape[:2]

print('img2 size: {} x {}'.format(w, h))
```

```python
if len(img1.shape) == 2: #img1.ndim ==2 해도 똑같다.

    #shape이라는 튜플값이 2이냐,

    print('img1 is a grayscale image')

elif len(img1.shape) == 3:

    print('img1 is a truecolor image')
```

```python
plt.subplot(121),plt.axis('off'),plt.imshow(img1,cmap='gray')
img2=cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
plt.subplot(122),plt.axis('off'),plt.imshow(img2)
```

```python
for y in range(h):

    for x in range(w):

        img1[y, x] = 255

        img2[y, x] = (0, 0, 255)
```

```python
plt.subplot(121),plt.axis('off'),plt.imshow(img1,cmap='gray')
img2=cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
plt.subplot(122),plt.axis('off'),plt.imshow(img2)
```

첫번째 이미지가 255임에도 검정색이었던 이유는, 255 한가지 색밖에 없기때문에 기준이 없어서 그렇다. 그래서 잘 나오게 하려면 vmin과 vmax를 사용해 주어야 한다.

```python
plt.subplot(121),plt.axis('off'),plt.imshow(img1,cmap='gray',vmin=0,vmax=255)
img2=cv2.cvtColor(img2,cv2.COLOR_BGR2RGB)
plt.subplot(122),plt.axis('off'),plt.imshow(img2)
```
