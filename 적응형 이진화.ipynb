OpenCV에서 제공하는 지역이진화의 한가지 방법입니다. 이전처럼 직접 4등분해서 구현할 필요가 없이 adaptiveMethod를 사용합니다.   
여기서 thresholdType은 그냥 이진화 할 것인지, 반전할 것인지를 선택할 수 있으며  
BlockSize에는 3이상의 홀수로 블록 크기를 지정해줍니다. C값은 임계값을 결정하는 파라미터가 되겠습니다.  

```python
import sys
import numpy as np
import cv2
import matplotlib.pyplot as plt

!pip install opencv-python
```

```python
src = cv2.imread('sudoku.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
if src is None:

    print('Image load failed!')

    sys.exit()
```

트랙바를 지정해 줍니다.

```python
def on_trackbar(pos):

    bsize = pos #3이상 홀수 이어야하기때문이다.

    if bsize % 2 == 0:

        bsize = bsize - 1

    if bsize < 3:

        bsize = 3
```

마지막에서 두번째 파라미터는 블럭 단위정도로 보시면됩니다. 숫자가 클수록 블럭을 작고 많이 나누게됩니다.

```python
 dst = cv2.adaptiveThreshold(src, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY, 129, 5)
```

```python
plt.subplot(121),plt.axis('off'),plt.imshow(src,cmap='gray')
plt.subplot(122),plt.axis('off'),plt.imshow(dst,cmap='gray')
```
