해밍 디스턴스는 2진수를 나열해서 비트가 서로다른게 몇개있는지 카운트한다. 따라서 실수 기술자는 L2Norm쓰고, 2진은 해밍디스턴스 실수기술자를 사용하는 알고리즘은 SIFT,SURF,KAZE가있고 2진기술자 사용하는 알고리즘은 AKAZE,ORB,BRIEF 가 있습니다.

```python
import sys
import numpy as np
import cv2
import matplotlib.pyplot as plt
import random

!pip install opencv-python
```

```python
src1 = cv2.imread('graf1.png', cv2.IMREAD_GRAYSCALE)

src2 = cv2.imread('graf3.png', cv2.IMREAD_GRAYSCALE)
```

```python
if src1 is None or src2 is None:

    print('Image load failed!')

    sys.exit()
```

특징점 알고리즘 객체를 생성합니다. KAZE, AKAZE, ORB 중에 선택합니다.

```python
feature = cv2.KAZE_create()

#feature = cv2.AKAZE_create()

#feature = cv2.ORB_create()
```

특징점 검출 및 기술자를 계산합니다.  
디텍트를 먼저한 뒤에 컴퓨터를 하는 방식입니다.

```python
kp1 = feature.detect(src1) 

_, desc1 = feature.compute(src1, kp1)
```

디텍트와 컴퓨트(특징점과 기술점 계산을) 한번에 하는 방식입니다.

```python
kp2, desc2 = feature.detectAndCompute(src2, None)
```

```python


print('desc1.shape:', desc1.shape)

print('desc1.dtype:', desc1.dtype)

print('desc2.shape:', desc2.shape)

print('desc2.dtype:', desc2.dtype)
```

```python
dst1 = cv2.drawKeypoints(src1, kp1, None,

                         flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)

dst2 = cv2.drawKeypoints(src2, kp2, None,

                         flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
```

```python
plt.axis('off'),plt.imshow(dst1)
```

```python
plt.axis('off'),plt.imshow(dst2)
```

```python

cv2.imshow('dst1', dst1)

cv2.imshow('dst2', dst2)

cv2.waitKey()

cv2.destroyAllWindows()
```
