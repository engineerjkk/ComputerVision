```python
import sys
import numpy as np
import cv2
import matplotlib.pyplot as plt

!pip install opencv-python
```

```python
src=cv2.imread('cells.png', cv2.IMREAD_GRAYSCALE)
```

```python
if src is None:
    print('Image load failed!')
    sys.exit()
```

 _, dst1 으로 인자 두개를 받습니다.  
 스레시 홀드는 100으로 지정해 줍니다.

```python
_,dst1=cv2.threshold(src,100,255,cv2.THRESH_BINARY)
```

```python
_, dst2=cv2.threshold(src,210,255,cv2.THRESH_BINARY)
```

```python
plt.axis('off')

plt.subplot(131),plt.axis('off'),plt.imshow(src,cmap='gray')
plt.subplot(132),plt.axis('off'),plt.imshow(dst1,cmap='gray')
plt.subplot(133),plt.axis('off'),plt.imshow(dst2,cmap='gray')
```

```python
cv2.imshow('src',src)
cv2.imshow('dst1',dst1)#염색세포 검출
cv2.imshow('dst2',dst2)#모든세포 검출
cv2.waitKey()
cv2.destroyAllWindows()
```
